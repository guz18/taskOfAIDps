```python
import pandas as pd
import matplotlib.pyplot as plt

# read the CSVs
df1 = pd.read_csv('predictions\\2prediction.csv', index_col=0)
df2 = pd.read_csv('predictions\\3prediction.csv', index_col=0)
df3 = pd.read_csv('predictions\\4prediction.csv', index_col=0)

# read the real values
actuals = df1['Actuals']

# make the plot

# Plot oluştur
plt.figure(figsize=(10, 5))

# Gerçek değerleri çiz
plt.plot(actuals.index, actuals.values, label='Actual', linewidth=2)

# Tahmin değerlerini çiz (ince ve noktalı çizgiler kullan)
plt.plot(df1.index, df1['Predictions'].values, label='Predictions of model 2', linestyle='dotted', linewidth=0.5)
plt.plot(df2.index, df2['Predictions'].values, label='Predictions of model 3', linestyle='dotted', linewidth=0.5)
plt.plot(df3.index, df3['Predictions'].values, label='Predictions of model 4', linestyle='dotted', linewidth=1.5)

plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Predicted vs Actual Values for 2021')
plt.legend()
plt.show()
```
